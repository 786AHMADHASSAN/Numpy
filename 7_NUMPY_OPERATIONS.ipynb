## Rules for dot and cross product

For dot product:
1. The dot product of two matrices is only defined when the number of columns in the first matrix is equal to the number of rows in the second matrix.
2. The resulting matrix from the dot product operation will have dimensions determined by the number of rows of the first matrix and the number of columns of the second matrix.

For cross product:
1. The cross product is only defined for vectors in three-dimensional space.
2. The result of the cross product is a vector that is orthogonal (perpendicular) to the two input vectors and has a magnitude equal to the area of the parallelogram formed by the two input vectors.
